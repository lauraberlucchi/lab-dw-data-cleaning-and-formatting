## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they contain only valid data. 

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
# Your code here
insurance_df['State'].unique()
```

```python
new_states = {'Washington':'Washington', 'Arizona':'Arizona', 'Nevada':'Nevada', 'California':'California', 'Oregon':'Oregon', 
              'Cali': 'California','AZ':'Arizona', 'WA':'Washington'}
insurance_df['State'] = insurance_df['State'].map(new_states)
insurance_df['State'].unique()

```

```python
insurance_df['Gender'].unique()
```

```python
gender_id = {'F':'F', 'M':'M', 'Femal':'F', 'Male':'M', 'female':'F'}
insurance_df['Gender'] = insurance_df['Gender'].map(gender_id)
insurance_df['Gender'].unique()
```

```python
insurance_df['Education'].unique()
```

```python
education_levels = {'Master':'Master', 'Bachelor':'Bachelor', 'High School or Below':'High School or below', 
                    'College':'Bachelor','Bachelors':'Bachelor', 'Doctor':'Doctor'}
insurance_df['Education'] = insurance_df['Education'].map(education_levels)
insurance_df['Education'].unique()
```

```python
insurance_df['Customer_Lifetime_Value'].info()
```

```python
#first I tried lambda x: x.strip(%) but I got an AttributeValue - though the dtype is object - because of nan values.
#so, I apply strip() only if x is not a nan value
insurance_copy_df = insurance_df.copy()
insurance_df['Customer_Lifetime_Value'] = insurance_df['Customer_Lifetime_Value'].apply(lambda x: str(x).strip('%') if not pd.isna(x) else x)
```

```python
insurance_df
```

```python
insurance_df['Vehicle_Class'].unique()
```

```python
new_classes = {'Four-Door Car':'Four-Door Car', 'Two-Door Car':'Two-Door Car', 'SUV': 'SUV',
               'Luxury SUV':'Luxury Car', 'Sports Car':'Luxury Car','Luxury Car':'Luxury Car'}
insurance_df['Vehicle_Class']=  insurance_df['Vehicle_Class'].map(new_classes)
insurance_df['Vehicle_Class'].unique()
```

```python
insurance_df
```

```python
#if I find that all 3 values in this columns are nan, I think it is safe to drop the row as useless
insurance_df = insurance_df.dropna(subset=['Customer','Monthly_Premium_Auto','Total_Claim_Amount'], how='all')
insurance_df
```

```python
insurance_df['Gender'].value_counts()
```

```python
# since the ration between males and females seems close to even, I will try to fill the nan values with ffill
insurance_df['Gender'] = insurance_df['Gender'].fillna(method='ffill')
```

```python
insurance_df['Gender'] = insurance_df['Gender'].fillna(method='bfill')
```

```python
insurance_df.isna().sum()
```

```python
#for the 3 nan values in Customer_Lifetime_Value, I am choosing to fill them with the median (as they are now float type)
insurance_df['Customer_Lifetime_Value'] = insurance_df['Customer_Lifetime_Value'].fillna(insurance_df['Customer_Lifetime_Value'].mean())
```

```python
insurance_df['Gender'].value_counts()
```

```python
insurance_df.isna().sum()
```

```python
#for Customer_Lifetime_Values i choose to fill the 3 nan position with the median
insurance_df['Customer_Lifetime_Value'].fillna(insurance_df['Customer_Lifetime_Value'].mean(), inplace=True)
```

```python
insurance_df.isna().sum()
```

```python
#finally: converting numeric data types in integers
insurance_df.dtypes
```
